```python
from zipfile import ZipFile
from urllib.request import urlopen   
import pandas as pd
import os

URLs = ['https://s3.amazonaws.com/tripdata/201801-citibike-tripdata.csv.zip',\
        'https://s3.amazonaws.com/tripdata/201802-citibike-tripdata.csv.zip',\
        'https://s3.amazonaws.com/tripdata/201803-citibike-tripdata.csv.zip',\
        'https://s3.amazonaws.com/tripdata/201804-citibike-tripdata.csv.zip',\
        'https://s3.amazonaws.com/tripdata/201805-citibike-tripdata.csv.zip']
       

# open and save the zip file onto computer
dfs=[]
for URL in URLs:
# open and save the zip file onto computer
    url = urlopen(URL)
    output = open('zipFile.zip', 'wb')    # note the flag:  "wb"        
    output.write(url.read())
    output.close()

# read the zip file as a pandas dataframe
    df = pd.read_csv('zipFile.zip')      
    dfs.append(df)   
```
